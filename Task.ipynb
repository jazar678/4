```
public interface IEmployee{
    string GetFullName();
    decimal GetSalary();

    string Position { get; set; }
}

public interface ISupplier{
    string GetName();
    decimal GetCost();

    string SupplierType { get; set; }
}

public interface IEnterprise {
    void AddSupplier(ISupplier supplier);
}

public interface IListSupplier{
    public void AddSupplier(ISupplier supplier);
    public void RemoveSupplier(ISupplier supplier);
}

public interface IListEmployee{
    public void AddEmployee(Employee employee);
    public void RemoveEmployee(Employee employee);
}

public interface IGetterList{
    public void getListSuppliers();
    public void getListEmployees();
}

public class Enterprise : IEnterprise, IListEmployee{

    List<ISupplier> suppliers = new List<ISupplier>();
    List<Employee> employees = new List<Employee>();

    private decimal TotalCost {get; set;}
    private decimal TotalSalary {get; set;}


    public void AddSupplier(ISupplier supplier){
        suppliers.Add(supplier);
        TotalCost += supplier.GetCost();
    }

    public void RemoveSupplier(ISupplier supplier){
        suppliers.Remove(supplier);
        TotalCost -= supplier.GetCost();

    }

    public void AddEmployee(Employee employee){
        employees.Add(employee);
        TotalSalary += employee.GetSalary();
    }

    public void RemoveEmployee(Employee employee){
        employees.Remove(employee);
        TotalSalary -= employee.GetSalary();
    }

    public void getListSuppliers(){
        ISupplier[] arr = suppliers.ToArray();

        Console.WriteLine("Список поставщиков:");

        for (int i = 0; i < arr.Length; i++){
            Console.WriteLine($"Имя: {arr[i].GetName()}. Стоимость услуг: {arr[i].GetCost()}");
        }

        Console.WriteLine($"Общие расходы на посатвщиков: {TotalCost}");
    }

    public void getListEmployees(){
        Employee[] arr = employees.ToArray();

        Console.WriteLine("Список сотрудников:");

        for (int i = 0; i < arr.Length; i++){
            Console.WriteLine($"Имя: {arr[i].GetFullName()}, Должность: {arr[i].GetPosition()}, Зарплата: {arr[i].GetSalary()}");
        }

        Console.WriteLine($"Общие расходы на сотрудников: {TotalSalary}");
    }

    public void getTotal(){
        Console.WriteLine( $"Общие расходы: {TotalSalary + TotalCost}");
    }
}

public class Supplier : ISupplier {
    public string SupplierType { get; set; }
    private string Name { get; set; }
    public decimal Cost { get; set; }

    public Supplier(string supplierType, string name, decimal cost){
        SupplierType = supplierType;
        Name = name;
        Cost = cost;
    } 

    public string GetName(){
        return Name;
    }
    public decimal GetCost(){
        return Cost;
    }
}

public class Employee : IEmployee {
    public string Position { get; set; }
    private string Name { get; set; }
    public decimal Salary { get; set; }

    public Employee(string position, string name, decimal salary){
        Salary = salary;
        Name = name;
        Position = position;
    } 

    public string GetFullName(){
        return Name;
    }
    
    public decimal GetSalary(){
        return Salary;
    }

    public string GetPosition(){
        return Position;
    }
}


Enterprise enterprise = new Enterprise();

Supplier supplier1 = new Supplier("Доставка компонентов ПК", "ООО 'Мираторг'", 75000);
Supplier supplier2 = new Supplier("Поставка еды в столовую", "ООО 'Агрогном'", 105000);

Employee employee1 = new Employee("Менеджер", "Иванов Иван Иванович", 65000);
Employee employee2 = new Employee("Программист", "Марфа Петровна Данильевна", 85000);

enterprise.AddEmployee(employee1);
enterprise.AddEmployee(employee2);

enterprise.AddSupplier(supplier1);
enterprise.AddSupplier(supplier2);

enterprise.getListSuppliers();

Console.WriteLine();

enterprise.getListEmployees();

Console.WriteLine();

enterprise.getTotal();
```
